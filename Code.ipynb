# Handling NaN values - Impute the training data set with the median using Simple Imputer

```python
# Check whether the missing values are filled or not
print("training_data missing values:")
print(training_data.isna().sum())

# If printed 0 for all columns, then there are no longer NaN values in the data set
print("\ntraining_data missing values overall amount:")
print(training_data.isna().sum().sum())
```

```python
"""
Fill the missing values in the data set using SimpleImputer which essentailly imputes the missing values with the mean.
"""

# Importing SimpleImputer which will allow us to fill the missing values with the mean
from sklearn.impute import SimpleImputer

# Set the imputer to mean
imputer = SimpleImputer(strategy='median')

# Impute missing values using the mean of the corresponding feature
imputed_training_data = imputer.fit_transform(training_data)
# Convert the imputed data back to Pandas DataFrame
imputed_training_data = pd.DataFrame(imputed_training_data, columns=training_data.columns)

# Check whether the missing values are filled or not
print("Imputed training_data missing values:")
print(imputed_training_data.isna().sum())

# If printed 0 for all columns, then there are no longer NaN values in the data set
print("\nImputed training_data missing values overall amount:")
print(imputed_training_data.isna().sum().sum())

# Preview the data again
print("\nImputed Data head:")
print(imputed_training_data.head())
print(imputed_training_data.shape)
```

# Testing different Feature Scaling algoritms using cross-validation on a winning classification model

* Min-Max
* Standard Scaling
* No-scaling

```python
"""
Evaluation and selection of a feature scaling method for the data.
"""
# Importing essential libraries
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC

# Split the training data into a training and validation set
X_train, X_val, y_train, y_val = train_test_split(imputed_training_data, target, test_size=0.2, random_state=42)

# Function that Normalisation
def min_max_scaling(X_train, X_val):
    scaler = MinMaxScaler()
    scaler.fit(X_train)
    return scaler.transform(X_train), scaler.transform(X_val)

# Function that Standardisation
def standard_scaling(X_train, X_val):
    scaler = StandardScaler()
    scaler.fit(X_train)
    return scaler.transform(X_train), scaler.transform(X_val)

X_train_mm, X_val_mm = min_max_scaling(X_train, X_val)
X_train_std, X_val_std = standard_scaling(X_train, X_val)

# Function that trains the data set and evaluates using an SVC model with an RBF kernel
def train_and_evaluate(X_train, y_train, X_val, y_val):
    model = SVC(random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)
    return accuracy_score(y_val, np.ravel(y_pred))

# Evaluates the model without any Scaling
accuracy_non_scaled = train_and_evaluate(X_train, y_train, X_val, y_val)

# Evaluates the model with Min-Max Scaling
accuracy_mm = train_and_evaluate(X_train_mm, y_train, X_val_mm, y_val)

# Evaluates the model with Standard Scaling
accuracy_std = train_and_evaluate(X_train_std, y_train, X_val_std, y_val)

# Printing the accuracies
print(f"Accuracy without any scaling: {accuracy_non_scaled}")
print(f"Accuracy with Min-Max Scaling: {accuracy_mm}")
print(f"Accuracy with Standard Scaling: {accuracy_std}")
```

# Tuning Hyperparameters using Gird Search


```python
# Import necessary libraries
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.svm import SVC

def evaluate_preds(y_true, y_preds):
  """
  Performs evaulutation comparison on y_true labels vs. y_pred labels on a classifiaction.
  """

  accuracy = accuracy_score(y_true, y_preds)
  precision = precision_score(y_true, y_preds, zero_division=0)
  recall = recall_score(y_true, y_preds)
  f1 = f1_score(y_true, y_preds)
  metric_dict = {"accuracy": round(accuracy, 2),
                 "precision": round(precision, 2),
                 "recall": round(recall, 2),
                 "f1": round(f1, 2)}
  print(f"Acc: {accuracy * 100:.2f}%")
  print(f"Precision: {precision:.2f}")
  print(f"Recall: {recall:.2f}")
  print(f"F1 score: {f1:.2f}")

  return metric_dict
```

```python
"""
Perform random search for hyperparameter tuning on an SVM classifier
"""

# Split the data into train and validation sets
X_train, X_valid, y_train, y_valid = train_test_split(final_training_data, target, test_size=0.2, random_state=42)

# Define the parameter grid for the grid search
param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid'],
    'degree': [2, 3, 4],
    'gamma': ['scale', 'auto', 0.1, 1, 10]
}

# Create an SVC object and the GridSearchCV object
svm = SVC(random_state=42)
grid_search = GridSearchCV(svm, param_grid=param_grid, cv=5, n_jobs=-1)

# Fit the GridSearchCV to the training data
grid_search.fit(X_train, np.ravel(y_train))

# Get the best parameters found by the grid search
best_params = grid_search.best_params_
print("Best Parameters: ", best_params)

# Train an SVM classifier with the best parameters and evaluate its performance
best_svm = SVC(**best_params, random_state=42)
best_svm.fit(X_train, np.ravel(y_train))

y_preds = best_svm.predict(X_valid)
svm_metrics = evaluate_preds(y_valid, y_preds)
svm_metrics
```
